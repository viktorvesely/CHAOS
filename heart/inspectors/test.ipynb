```python
import cell
import numpy as np
import matplotlib.pyplot as plt
```

```python
V = np.arange(-90, -20, 0.5)
alpha_m = cell.alpha_m(V)
beta_m = cell.beta_m(V)
alpha_h = cell.alpha_h(V)
beta_h = cell.beta_h(V)
m_inft = cell.steady_state(alpha_m, beta_m)
h_inft = cell.steady_state(alpha_h, beta_h)
m_tau = cell.tau(alpha_m, beta_m)
h_tau = cell.tau(alpha_h, beta_h)

plt.plot(V, m_tau, label="m-tau")
plt.plot(V, h_tau, label="h-tau")
plt.legend()
plt.show()
```

```python
plt.plot(V, m_inft, label="m-inft")
plt.plot(V, h_inft, label="h-inft")
plt.legend()
plt.show()
```

```python
import numpy as np


def kahan(vec):
    acc = 0
    c = 0

    for el in vec:
        small = el - c
        temp = acc + small
        c = (temp - acc) - small
        acc = temp
    
    return acc


test = np.random.random(5_000_000)

k_sum = kahan(test)
n_sum = np.sum(test)

print(k_sum)
print(n_sum)

print(k_sum - n_sum)
```
