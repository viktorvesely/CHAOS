```python
import cell
import numpy as np
import matplotlib.pyplot as plt
```

```python
V = np.arange(-90, -20, 0.5)
alpha_m = cell.alpha_m(V)
beta_m = cell.beta_m(V)
alpha_h = cell.alpha_h(V)
beta_h = cell.beta_h(V)
m_inft = cell.steady_state(alpha_m, beta_m)
h_inft = cell.steady_state(alpha_h, beta_h)
m_tau = cell.tau(alpha_m, beta_m)
h_tau = cell.tau(alpha_h, beta_h)

plt.plot(V, m_tau, label="m-tau")
plt.plot(V, h_tau, label="h-tau")
plt.legend()
plt.show()
```

```python
plt.plot(V, m_inft, label="m-inft")
plt.legend()
plt.show()
```

```python
V = np.arange(-90, 40, 0.5)
alpha_X = cell.alpha_X(V)
beta_X = cell.beta_X(V)
X_inft = cell.steady_state(alpha_X, beta_X)
X_tau = cell.tau(alpha_X, beta_X)

plt.plot(V, X_inft, label="inft")
plt.plot(V, X_tau, label="tau")
plt.legend()
plt.show()
```

```python
plt.plot(V, alpha_X, label="alpha")
plt.plot(V, beta_X, label="beta")
plt.legend()
```

```python
V = np.arange(-90, 40, 0.5)
alpha = cell.alpha_h(V)
beta = cell.beta_h(V)
inft = cell.steady_state(alpha, beta)
tau = cell.tau(alpha, beta)

plt.plot(V, inft, label="inft")
plt.plot(V, tau, label="tau")
plt.legend()
plt.show()
```
